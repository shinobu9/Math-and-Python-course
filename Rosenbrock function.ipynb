```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import linalg, interpolate, optimize

%matplotlib inline
```

```python
def rosenbrock(x):
    return (1 - x[0])**2 + 10.0*(x[1] - x[0]**2)**2
print(rosenbrock([1,1]))
```

```python
result= optimize.brute(rosenbrock, ((-5,5),(-5,5)))
result
```

```python
%matplotlib notebook

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d   

x = np.arange(-2, 2, 0.15)
y = np.arange(-1, 3, 0.15)
x, y = np.meshgrid(x, y)
z = rosenbrock([x,y])

fig = plt.figure(figsize=(12,7))
ax = fig.gca(projection='3d')
#ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x,y,z, cmap='gist_heat_r', linewidth=0, antialiased=False)
ax.plot(result[0],result[1],rosenbrock(result), color='green', marker='*', markersize=25)
ax.set_zlim(0, 200)
fig.colorbar(surf, shrink=0.5, aspect=10)

```
